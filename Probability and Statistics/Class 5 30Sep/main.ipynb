### Exploratory Data Analysis:

```python
df.columns
```

#### Gender

```python
ax = sns.countplot(x = 'Gender', data = df);
```

```python
ax = sns.countplot(x = 'Gender', data = df, hue='Gender');

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum()
sales_gen
```

```python
sns.barplot(x= 'Gender', y = 'Amount', data= sales_gen, hue='Gender')
```

**Conclusion**: Most of the customers are Female, Also the purchasing power of Female is higher than male according to the given data and corresponding plots

#### Age:

```python
ax = sns.countplot(x = 'Age Group', data = df, hue='Gender');

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum()
sales_age
```

```python
sns.barplot(x= 'Age Group', y = 'Amount', data= sales_age)
```

**Conclusion:** Most of the customers are in the age group of 26-35 Female

#### State:

```python
# Top 10 states with most number of Orders

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum()
sales_state = sales_state.sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State', y='Orders', data=sales_state)
```

```python
# Top 10 states with highest total amount

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum()
sales_state = sales_state.sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State', y='Amount', data=sales_state)
```

**Conclusion:** The Highest Number of Orders and also the highest order amount are from UP, Maharashtra, Karnataka.

#### Marital Status:

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(6,5)})

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum()

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

**Conclusion:** most of the buyers are unmarried (women) and they have high purchasing power

#### Occupation:

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum()

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

**Conlcusion:** most of the buyers are working in IT, Healthcare and Aviation sector

#### Product Category:

```python
sns.set(rc={'figure.figsize':(25,6)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

**Conclusion:** most of the sold products are from Food, Clothing and Electronics category

#### Product ID:

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

##### **Conclusion:** Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category

```python
# Zonal Analysis 

sales_zone = df.groupby('Zone').agg({'Orders': 'sum', 'Amount': 'sum'}).reset_index()
sns.barplot(x='Zone', y='Orders', data=sales_zone) # on Orders basis
# sns.barplot(x='Zone', y='Amount', data=sales_zone) #On amount basis
```

```python
# Age & Gender vs. Amount
sales_gender_age = df.groupby(['Gender', 'Age Group']).agg({'Amount': 'sum', 'Orders': 'sum'}).reset_index()
sns.barplot(x='Age Group', y='Amount', hue='Gender', data=sales_gender_age)
```
