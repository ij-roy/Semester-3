```python
# imports
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('student_sleep_patterns.csv', encoding = 'unicode_escape')
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sleep_Duration', y='Sleep_Quality', data=df, hue='Gender', palette='viridis')
plt.title('Relationship Between Sleep Duration and Sleep Quality')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Sleep Quality (1-10)')
plt.grid(True)
plt.legend(title='Gender')
plt.show()
```

```python
ax = sns.countplot(x = 'Gender', data = df);
```

```python

df = df.dropna(subset=['Sleep_Duration', 'Sleep_Quality']) 

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sleep_Duration', y='Sleep_Quality', data=df)
plt.title('Relationship Between Sleep Duration and Sleep Quality')
plt.xlabel('Sleep Duration')
plt.ylabel('Sleep Quality')
# plt.grid(True)
plt.show()
```

```python
from scipy.stats import ttest_1samp


screen_time = df['Screen_Time'].dropna()

t_stat, p_value = ttest_1samp(screen_time, 5)

print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis:")
```
